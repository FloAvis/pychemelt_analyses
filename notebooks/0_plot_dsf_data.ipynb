Notebook showing how to import and plot protein unfolding data for combined thermal and chemical denaturation

```python
import pychemelt as pychem
import numpy as np
```

```python
dsf_file = "../data/nDSFdemoFile.xlsx"
```

```python
sample = pychem.Sample()
sample.read_multiple_files(dsf_file)

n_conditions = len(sample.conditions)

print('The total number of conditions is: {}'.format(n_conditions))
```

```python
concentrations = [
    0.05, 0.52, 1.0, 1.47, 1.95, 2.38, 2.5, 2.61, 2.73, 2.85, 2.97, 3.09,
    3.21, 3.33, 3.45, 3.56, 3.68, 4.25, 4.82, 5.39, 5.97, 6.54,
    np.nan, np.nan,
    0.05, 0.44, 0.83, 1.22, 1.61, 2.0, 2.39, 2.78, 3.17, 3.56, 3.95, 4.34,
    4.73, 5.12, 5.51, 5.9, 6.29, 6.68, 7.07, 7.46, 7.85, 8.24,
    np.nan, np.nan
]

sample.set_denaturant_concentrations(concentrations)
```

```python
conditions_of_interest =    [True for _ in range(8)] + [False for _ in range(48 - 8)])
```

```python
sample.set_signal(['350nm'])
sample.set_temperature_range(20,100)
```

```python
fig = pychem.plot_unfolding(sample)
fig.show()
```
